> **NumPy** ： 用于科学计算的基础库，提供了高性能的多维数组对象和用于处理这些数组的工具。 [quickstart](https://numpy.org/devdocs/user/quickstart.html)

> **Flask** 和 **Django** ： 两个流行的 Web 框架，用于构建 Web 应用程序。

> **SQLAlchemy** ： SQL 工具和对象关系映射（ORM）库，用于数据库操作。[sqlalchemy.org](https://www.sqlalchemy.org/)

> **OpenCV** ： 开源计算机视觉库，提供了丰富的图像处理和计算机视觉工具。

> **NLTK** 和 **SpaCy** ： 自然语言处理（NLP）库，用于文本处理和分析。

> **Pygame** ： 用于游戏开发的库，提供了多媒体功能和游戏开发工具。

> **Tkinter** 和 **PyQt** ： 用于创建图形用户界面（GUI）的库，分别是 Python 的内置库和基于 Qt 的库。

# NumPy 
是一个强大的 Python 库，用于科学计算和数据分析。它提供了许多用于数组操作和数学运算的函数。以下是 NumPy 中用于数据分析的一些常用函数和属性：

数组创建和操作：
np.array()：将列表或其他可迭代对象转换为数组。
np.zeros()：创建全零数组。
np.ones()：创建全一数组。
np.arange()：创建等差数组。
np.linspace()：创建等间隔数组。
np.reshape()：改变数组形状。
np.concatenate()：数组拼接。
np.split()：数组拆分。
np.sort()：对数组进行排序。
np.unique()：找到数组中的唯一值。
np.concatenate()：数组拼接。
np.vstack() 和 np.hstack()：垂直和水平堆叠数组。
np.reshape()：改变数组的形状。
np.ravel()：将多维数组转为一维数组。
np.newaxis 或 None：在数组中增加新维度。
np.flip()：沿指定轴翻转数组。

数组迭代：
np.nditer()：多维数组的迭代器。
np.ndenumerate()：多维数组的元素和索引的迭代器。

数学运算：
np.sum()：计算数组元素的和。
np.mean()：计算数组元素的平均值。
np.median()：计算数组元素的中位数。
np.std()：计算数组元素的标准差。
np.var()：计算数组元素的方差。
np.min()：找到数组中的最小值。
np.max()：找到数组中的最大值。
np.argmin()：找到最小值的索引。
np.argmax()：找到最大值的索引。
np.cumsum()：计算累积和。
np.cumprod()：计算累积积。

数组统计和线性代数：
np.histogram()：生成直方图。
np.bincount()：计算非负整数数组中每个值的频数。
np.corrcoef()：计算相关系数矩阵。
np.cov()：计算协方差矩阵。
np.dot()：矩阵乘法。
np.transpose() 或 array.T：转置数组。

统计量：
np.mean()：计算数组元素的平均值。
np.median()：计算数组元素的中位数。
np.std()：计算数组元素的标准差。
np.var()：计算数组元素的方差。
np.percentile()：计算百分位数。

随机数生成：
np.random.rand()：生成均匀分布的随机数。
np.random.randn()：生成标准正态分布的随机数。
np.random.randint()：生成整数随机数。
np.random.seed()：设置随机数种子。

数组索引和切片：
array[index]：通过索引获取数组元素。
array[start:stop]：切片获取数组的子集。
array[condition]：使用条件进行索引。

文件操作：
np.save() 和 np.load()：保存和加载数组数据。

# Pandas 
提供了一些其他用于数据分析的函数，可以提供有关数据集的描述性统计信息。以下是其中一些常用的函数：

Pandas DataFrame 提供了丰富的功能和操作，以下是其中一些常用的功能和方法：

创建 DataFrame：
pd.DataFrame(data)：从字典、列表、NumPy 数组等数据结构创建 DataFrame。
pd.read_csv()、pd.read_excel()：从文件中读取数据创建 DataFrame。

查看和了解数据：
df.head(n)、df.tail(n)：查看 DataFrame 的前 n 行或后 n 行。
df.info()：查看 DataFrame 的基本信息。
**df.describe()：生成关于 DataFrame 列的统计描述。**
**df.shape：获取 DataFrame 的行数和列数。**

索引和选择数据：
df['column_name']：选择单列。
df[['col1', 'col2']]：选择多列。
df.loc[row_indexer, col_indexer]：通过标签选择数据。
df.iloc[row_indexer, col_indexer]：通过整数位置选择数据。
**条件选择：df[df['column'] > value]。**

数据处理：
缺失值处理：df.dropna()、df.fillna(value)。
重复值处理：df.drop_duplicates()。
列重命名：df.rename(columns={'old_name': 'new_name'})。

数据排序：
df.sort_values(by='column', ascending=True)：按列值排序。

数据过滤和查询：
df.query('condition')：根据条件查询数据。

数据聚合和分组：
df.groupby('column').mean()：按列分组并计算均值。
df.groupby(['col1', 'col2']).agg({'col3': 'sum', 'col4': 'mean'})：多列分组并进行不同的聚合操作。

合并和连接数据：
pd.concat([df1, df2], axis=0)：纵向连接。
pd.concat([df1, df2], axis=1)：横向连接。
pd.merge(df1, df2, on='key')：根据列进行合并。

数据透视表：
pd.pivot_table(df, values='value', index='index_col', columns='col_to_pivot')。

时间序列操作：
pd.to_datetime(df['date_col'])：将列转换为日期时间类型。
df.resample('D').sum()：对日期时间列进行重新采样。

绘图与可视化：
df.plot()：绘制线图。
df.plot.bar()：绘制柱状图。
df.plot.scatter(x='col1', y='col2')：绘制散点图。

导出数据：
df.to_csv('filename.csv', index=False)：导出数据到 CSV 文件。
df.to_excel('filename.xlsx', index=False)：导出数据到 Excel 文件。

# Matplotlib 
是一个强大的绘图库，支持各种不同类型的图表。以下是一些常见的图表类型以及每种类型的简单示例：

```python
import matplotlib.pyplot as plt
import numpy as np
```

**折线图（Line Plot）：**

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

**散点图（Scatter Plot）：**

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

**柱状图（Bar Chart）：**

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

plt.bar(categories, values)
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

**直方图（Histogram）：**

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

**饼图（Pie Chart）：**

```python
labels = ['Category A', 'Category B', 'Category C']
sizes = [40, 30, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()
```

**箱线图（Box Plot）：**

```python
data = np.random.randn(100, 5)

plt.boxplot(data, labels=['A', 'B', 'C', 'D', 'E'])
plt.title('Box Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

**面积图（Area Plot）：**

```python
x = np.linspace(0, 1, 100)
y1 = x
y2 = x**2

plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4, label='Area between y1 and y2')
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')

plt.title('Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

**等高线图（Contour Plot）：**

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.title('Contour Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

**3D 散点图（3D Scatter Plot）：**

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)

ax.set_title('3D Scatter Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```
